```


def multbyalpha(b, f):
  t = b << 1
  n = len(bin(f)) - 3
  if t & (1<<n) != 0:
    return t^f
  else:
    return t
def multiplie(b, c, f):
  res = 0
  t = b
  k = c
  while t != 0:
    if (t & 1) != 0:
      res ^= k
    k = multbyalpha(k,f)
    t = t >> 1
  return res
def matmat_f(A,B,P):
  C = []
  for i in range(len(A)):
    L = []
    for j in range(len(B[0])):
      res = 0
      for k in range(len(B)):
        res ^= multiplie(A[i][k],B[k][j],P)
      L.append(res)
    C.append(L)
  return C
print(
  matmat_f([[14, 8, 9], [0, 4, 12], [12, 10, 12]],[[11, 12, 8], [7, 8, 15], [4, 9, 5]],19)
)
```
